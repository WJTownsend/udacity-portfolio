# The Goodest Bois: Conclusions from @WeRateDogs

Over the last few days, I've pulled an extensive amount of information from the tweets of [@WeRateDogs](https://twitter.com/dog_rates/), a lighthearted account which rates all manner of floofers, puppos, and doggos. Ratings are presented as being "out of 10", but the overwhelming majority are actually scored from 10 - 14 because [they're good dogs, Brent!](https://knowyourmeme.com/memes/theyre-good-dogs-brent) Data was pulled using information from Udacity and the Twitter API itself, looking specifically at the timeframe from mid-November 2015 to start of Aug 2017 (a bit over 2100 ratings), to answer three main questions:
- Do puppies or full grown dogs get higher ratings, on average?
- Do puppies or full grown dogs tend to get greater engagement on Twitter?
- What is the relationship between engagement with a tweet (retweets, favorites) and the rating issued to a dog?

## Pups vs Dogs: Who gets the better ratings? 

Using identifiers for puppies (puppo, pupper) and grown dogs (floofer, doggo), I was able to group tweets by whether they were for a puppo or a doggo. This grouping allows for an examination of how the cutest puppos or the goodest doggos are rated by @WeRateDogs:

Puppies have much much more scores than dogs, but I was surprised to see that the dogs have a higher average rating (10.91 for pups, 11.78 for dogs) with a tighter variance (2.83 for pups, 2.28 for dogs). Specifically, dogs seem to have a higher likelihood of scoring a 14 and a lower likelihood of scoring below a 10, relative to their pups. The 14's are often justified by the actions of a dog, which a pup would have a harder time achieving. For example, the first 14 was a dog who saved another dog from drowning, while the second was one of the search dogs looking for survivors in the wreckage of the World Trade Center after 9/11. The lower ratings were a surprise, and more analysis would be needed to determine why the smol dogs seem to get smol ratings.

## Pups vs Dogs: Who gets better engagement?

With pupperinos and woofbois already grouped out, I was also able to look at whether there was a difference in engagement on Twitter (retweets and favorites) between the two groups:

Looking at the mean engagement metrics for both pups and dogs, it's clear that the dogs are getting about twice as much engagement on Twitter than the puppies are. If we look at the median engagement as another measure of center, we see the dogs pull further ahead:

Having found that dogs get higher ratings on average than the pups, I did expect them to have higher engagement as well. That their engagement was this much higher, 2-2.5x that of the pups, was surprising. I thought that maybe some particularly large outliers in terms of engagement might be overstating this effect, but the effect actually grew stronger when using the median instead, which should mitigate the impact of those outliers. It seems that, surprisingly, the internet likes dogs more than they do puppies. 

## How do ratings impact Twitter engagement?

I also was curious to examine the relationship between a rating issued by @WeRateDogs and the resulting engagement from other Twitter users. A scatter plot was used to map out each tweet's engagement (retweets + favorites) relative to its rating:

We do see an overall trend that at the higher ratings, engagement tends to be much higher. The linear relationship here is not very strong, however, with a correlation coefficient of only 0.37. This makes sense, as we do see multiple local peaks (0, 5, 10, 13) and valleys (3, 8, 11, 14). I had considered that engagement might not be linear and might instead be quadratic in nature, with a parabola being the best fit. This was because social media tends to be a land of extremes: the internet seems to love elevating the best things and "dunking on" the worst things (and inadvertently elevating them as well). With this in mind, I tried plotting a linear, quadratic, and cubic best fit line to see if any of them seemed to fit the data better than the others:

It is immediately clear that the quadratic and cubic fits are much more accurate than a basic linear fit. The cubic fit appears to be closer than the quadratic at the high end (ratings of 14, in particular), but the quadratic fit is much closer than the cubic at the low end (ratings of 0, maybe 1 as well). This might lead to a conclusion that the quadratic line has an overall better fit here, but the small amount of data at the low end would mitigate the higher performance. As a result, I would say the cubic fit is probably the most accurate, and that the relationship between dog rating and tweet engagement is non-linear. 

## The Doggie Bag

So what can we take away from this data? As mentioned, the internet (or at least, Twitter followers of @WeRateDogs) seem to like grown dogs more than they like puppies. This was quite a surprise, as "cute" and "funny" tend to be great drivers of engagement on social media, perhaps only behind "anger" and "outrage". While I think puppies are generally cuter (and often funnier), perhaps I am in the minority here. It's also worth pointing out that, at least by @WeRateDogs criteria for issuing ratings, the highest ratings often require acts by the dog in question that elevate beyond cuteness. Pups lack vocational training and instead try to skate by on their youthful good looks, something that is undeniably adorable but fleeting in the long run. If puppers want to raise their overall perception on the internet, they clearly need to start working harder. 

Regarding the relationship between engagement and dog ratings, the non-linear nature of this wasn't particularly surprising. There is a truism that ["Each day on Twitter, there is one main character. The goal is to never be it."](https://twitter.com/maplecocaine/status/1080665226410889217) which refers to the fact that there is generally either someone being absolutely loved or being absolutely hated each day on social media, with really no in between. When looking at a scale of ratings, this obviously relates to how the extreme ends of the scale (low *or* high) are more interesting or evoking a stronger reaction than the mushy middle of things which are merely "fine". What was more interesting about this was that the cubic best fit line seemed to approximate the real data better than the quadratic one, implying a local peak in the middle of the rating scale. A larger and more robust sample size at the lower end of the ratings scale would be necessary to determine whether such a peak exists or if the quadratic best fit line would be a better approximation of the whole.
